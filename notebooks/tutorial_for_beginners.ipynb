## PART 3: What is a Neural Network?

It's like a chain of decisions:

```
Is there a stroke here?
        â†“
Are the strokes curved or angular?
        â†“
What's the overall shape?
        â†“
What script is this?
```

````python
# Let's manually simulate what a neural network does

# Step 1: Get pixel values from image
image = devanagari_char.flatten()  # Convert 2D to 1D
print(f"Image as input: {image}")
print(f"Number of inputs: {len(image)}")

# Step 2: Weights (the "brain" learns these)
# Each weight multiplies an input value
weights = np.random.randn(len(image)) * 0.1  # Random weights initially
print(f"\nNumber of weights: {len(weights)}")
print(f"Example weights (first 5): {weights[:5]}")

# Step 3: Calculate output
output = np.sum(image * weights)  # Multiply inputs by weights and sum
print(f"\nRaw output: {output}")

# Step 4: Convert to probability (0-1)
import math
probability = 1 / (1 + math.exp(-output/1000))  # Sigmoid function
print(f"Probability (0-1): {probability}")

print("\n" + "="*50)
print("That's the CORE of how neural networks work!")
print("="*50)
````

## PART 4: How the Model Learns

### Training Loop (Simplified)

````python
# Simulate training

class SimpleClassifier:
    """A toy neural network to understand learning"""
    
    def __init__(self, num_features):
        self.weights = np.random.randn(num_features) * 0.01
        self.bias = 0
        self.loss_history = []
    
    def predict(self, X):
        """Make a prediction"""
        return X @ self.weights + self.bias
    
    def train(self, X, y, epochs=10, learning_rate=0.01):
        """Train the model"""
        for epoch in range(epochs):
            # Forward pass: make prediction
            predictions = self.predict(X)
            
            # Calculate error
            error = predictions - y
            loss = np.mean(error ** 2)  # Mean squared error
            self.loss_history.append(loss)
            
            # Backward pass: calculate gradients
            dW = 2 * X.T @ error / len(X)
            db = 2 * np.mean(error)
            
            # Update weights (this is learning!)
            self.weights -= learning_rate * dW
            self.bias -= learning_rate * db
            
            if (epoch + 1) % 2 == 0:
                print(f"Epoch {epoch+1}: Loss = {loss:.4f}")

# Create training data
X_train = np.array([
    [255, 100, 100, 100, 255, 255, 100, 200, 100, 255],  # Devanagari-like
    [255, 100, 100, 100, 255, 255, 100, 200, 100, 255],
    [255, 200, 100, 200, 255, 255, 200, 100, 200, 255],  # Tamil-like
    [255, 200, 100, 200, 255, 255, 200, 100, 200, 255]
]) / 255.0  # Normalize

y_train = np.array([0, 0, 1, 1])  # 0=Devanagari, 1=Tamil

# Train the model
model = SimpleClassifier(num_features=X_train.shape[1])
print("Training in progress...\n")
model.train(X_train, y_train, epochs=10, learning_rate=0.1)

# Plot loss over time
plt.figure(figsize=(10, 4))
plt.plot(model.loss_history, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('How Loss Decreases During Training')
plt.grid(True, alpha=0.3)
plt.show()

print("\n" + "="*50)
print("See how loss decreases?")
print("That means the model is LEARNING!")
print("="*50)
````

## PART 5: Evaluating Your Model

How do we know if it's good?

````python
# After training, test on new data
X_test = np.array([
    [255, 100, 100, 100, 255, 255, 100, 200, 100, 255],  # New Devanagari
    [255, 200, 100, 200, 255, 255, 200, 100, 200, 255],  # New Tamil
]) / 255.0

y_test = np.array([0, 1])  # Actual labels

# Make predictions
predictions = model.predict(X_test)
predicted_labels = (predictions > 0.5).astype(int)  # Convert to 0 or 1

# Calculate accuracy
accuracy = np.mean(predicted_labels == y_test)

print("Test Results:")
print(f"True labels:      {y_test}")
print(f"Predictions:      {predictions.round(3)}")
print(f"Predicted labels: {predicted_labels}")
print(f"\nAccuracy: {accuracy*100:.0f}%")

if accuracy == 1.0:
    print("\nâœ“ Perfect! The model learned correctly!")
elif accuracy >= 0.8:
    print("\nâœ“ Good! The model is working well!")
elif accuracy >= 0.5:
    print("\nâš  Okay, but needs improvement")
else:
    print("\nâœ— Not good, need more training or data")
````

## PART 6: Understanding Your Project Code

Now let's look at your actual preprocessing code!

````python
import sys
sys.path.append('..')

from src.preprocessing import ImagePreprocessor

# Initialize the preprocessor
preprocessor = ImagePreprocessor(
    target_height=64,
    target_width=256,
    binarization_method='adaptive'
)

# Create a sample noisy image
sample_image = np.ones((100, 200), dtype=np.uint8) * 200
# Add some "text" (simple pattern)
sample_image[40:60, 50:80] = 50  # Dark rectangle (text)
sample_image[40:60, 100:130] = 50  # Another rectangle

# Add noise
noise = np.random.randint(0, 30, sample_image.shape)
noisy_image = sample_image + noise
noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)

# Show preprocessing steps
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Original
axes[0, 0].imshow(noisy_image, cmap='gray')
axes[0, 0].set_title('1. Original (Noisy)')
axes[0, 0].axis('off')

# Denoised
denoised = preprocessor.remove_noise(noisy_image)
axes[0, 1].imshow(denoised, cmap='gray')
axes[0, 1].set_title('2. After Denoising')
axes[0, 1].axis('off')

# Contrast enhanced
enhanced = preprocessor.enhance_contrast(denoised)
axes[0, 2].imshow(enhanced, cmap='gray')
axes[0, 2].set_title('3. After Contrast Enhancement')
axes[0, 2].axis('off')

# Binarized
binary = preprocessor.apply_binarization(enhanced)
axes[1, 0].imshow(binary, cmap='gray')
axes[1, 0].set_title('4. After Binarization')
axes[1, 0].axis('off')

# Resized
resized = preprocessor.resize_image(binary)
axes[1, 1].imshow(resized, cmap='gray')
axes[1, 1].set_title('5. After Resizing')
axes[1, 1].axis('off')

# Normalized
normalized = preprocessor.normalize(resized)
axes[1, 2].imshow(normalized, cmap='gray')
axes[1, 2].set_title('6. After Normalization')
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()

print("\nPreprocessing Pipeline Explained:")
print(f"1. Original shape: {noisy_image.shape}")
print(f"2. Denoised shape: {denoised.shape}")
print(f"3. Enhanced shape: {enhanced.shape}")
print(f"4. Binary shape: {binary.shape}")
print(f"5. Resized shape: {resized.shape}")
print(f"6. Normalized dtype: {normalized.dtype} (values 0-1)")
print("\nWHY each step?")
print("  1. Remove noise â†’ Clean image")
print("  2. Enhance contrast â†’ Make text darker, background lighter")
print("  3. Binarize â†’ Pure black and white (easier for model)")
print("  4. Resize â†’ Standard size (224x224) for model")
print("  5. Normalize â†’ Values 0-1 (better for neural networks)")
````

## PART 8: Hands-On: Experiment with Parameters

Change values and see what happens!

````python
# Let's train the model with DIFFERENT learning rates and see the difference

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

learning_rates = [0.01, 0.1, 1.0]

for idx, lr in enumerate(learning_rates):
    model = SimpleClassifier(num_features=X_train.shape[1])
    
    for epoch in range(20):
        predictions = model.predict(X_train)
        error = predictions - y_train
        loss = np.mean(error ** 2)
        model.loss_history.append(loss)
        
        dW = 2 * X_train.T @ error / len(X_train)
        db = 2 * np.mean(error)
        
        model.weights -= lr * dW
        model.bias -= lr * db
    
    axes[idx].plot(model.loss_history, marker='o')
    axes[idx].set_title(f'Learning Rate = {lr}')
    axes[idx].set_xlabel('Epoch')
    axes[idx].set_ylabel('Loss')
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nObservations:")
print("  - Learning Rate 0.01: Slow improvement (small steps)")
print("  - Learning Rate 0.1: Good improvement (balanced)")
print("  - Learning Rate 1.0: Unstable (too big steps)")
print("\nThis shows why choosing the RIGHT learning rate is important!")
````
